### Numerical Time Series - features extraction 
Using [statsmodel.OLS] (http://statsmodels.sourceforge.net/0.5.0/generated/statsmodels.regression.linear_model.OLS.html)
for fitting a regression on the transactions and identifying outliers.

<H3>Adding feedback from agents </H3>

```python
def agent_feedback(row):
    if row.outliers>0 and row.nlp_fraud_score>0.0 and row.content_topic>0 and row.fr_degree_centrality<0.02 \
     and row.fr_graph_connections<500:
        return 1
    else:
        return 0
cdf['agent_feedback']=cdf.apply(agent_feedback, axis =1)
display(cdf.head(5))
plt.figure(figsize = (7,5))
viz.plot_corr(cdf[['amount','transactions','big_trans','is_in_odd_day','is_at_odd_time','outliers','subject_sentiment','content_sentiment', 
                   'subject_power','content_power','subject_topic','content_topic','nlp_fraud_score',
                   'fr_degree_centrality','fr_graph_connections','agent_feedback']], cmap='RdBu', mask_color='gray')
```

### Modeling and Evaluation

Splitting data into train and test subsets and fitting an SGD classifier and a Random Forest classifier for learning the agents' feedback and comapring the performance.


```python
feature_names = ['amount','transactions','big_trans','is_in_odd_day','is_at_odd_time','outliers','subject_sentiment','content_sentiment', 
                   'subject_power','content_power','subject_topic','content_topic','nlp_fraud_score',
                   'fr_degree_centrality','fr_graph_connections']
ground_truth_field =['agent_feedback']
feature_vectors = cdf[feature_names]
target_vector = cdf[ground_truth_field]
split_point = int(round(0.8*len(feature_vectors)))

train_features = DataFramePlus(feature_vectors[:split_point])
train_target = DataFramePlus(target_vector[:split_point])
test_features = DataFramePlus(feature_vectors[split_point:])
test_target = DataFramePlus(target_vector[split_point:])
```

```python
%%time
import xpatterns.analytics.modeling.model 
from xpatterns.analytics.modeling.model import *
sgd = Model('sgd_classifier',model_type = SKLearnTypes.SGDClassifier)
sgd.submodel.loss='log'
sgd.fit(train_features, ravel(train_target))
sgd_test_predicted_proba = sgd.predict_proba(test_features)

model_kwargs={'min_samples_leaf':1,}
dt = Model('random forest',model_type = SKLearnTypes.RandomForestClassifier, model_kwargs=model_kwargs)
dt.fit(train_features, ravel(train_target))
dt_test_predicted_proba = dt.predict_proba(test_features)
```

The *plot_roc_curve()* function can be used to plot a [Receiver Operating Characteristic (ROC)](https://en.wikipedia.org/wiki/Receiver_operating_characteristic) curve and compute the [Area Under the Curve (AUC)](https://en.wikipedia.org/wiki/Receiver_operating_characteristic#Area_under_the_curve), a number between 0 and 1 that can be used as a single-valued metric.

```python
viz.plot_roc_curves(test_target.values, 
                    sgd_test_predicted_proba[:,1],'SGD',
                    dt_test_predicted_proba[:,1], 'RF')
```

### Model Optimization

We can use the [Grid Search] to automate the search for the best parameters for modeling this data, using 5-fold cross validation (*cv=5*) and AUC as a scoring metric (*scoring='roc_auc'*).

We can specify possible parameter values for any parameters used by a model. A [RandomForestClassifier model](http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html) has a 11 parameters. We'll simplify the search by only selecting 2 - *criterion* and *min_samples_leaf* - to optimize.

* *min_samples_leaf* : The minimum number of samples in newly created leaves. A split is discarded if after the split, one of the leaves would contain less then min_samples_leaf samples.
* *criterion* : The criterion to be used when splitting at each node (*gini* or *entropy*)

We'll focus the search on determining whether or not to *bootstrap* and whether to use 100, 10 or 2 *min_samples_leaf*:

```python
%%time
train_target_values = []
for v in train_target.values:
    train_target_values.append(int(v))
from sklearn.grid_search import GridSearchCV
params = {
    'min_samples_leaf':[1,2,3,4,5,10,50,100],
    'criterion':['gini','entropy']
    }
gs = GridSearchCV(dt.submodel, params, cv=5, scoring='roc_auc')
gs.fit(train_features.values, train_target_values)
```
