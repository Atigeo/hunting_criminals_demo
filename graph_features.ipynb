<h3>Graph(s) of the people's communications (subset of 1000 edges)</h3>

```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph() # create empty graph

for gr,vals in df[:1000].groupby(['fr','to']):
    G.add_edge(gr[0],gr[1],message=len(vals))

try: 
    pos = nx.graphviz_layout(G)
    plt.figure(figsize=(50,50))
    nx.draw(G,pos,node_size=10,node_color='red',alpha=1,edge_color='orange',font_size=16)
    nx.draw_networkx_nodes (G,pos,alpha=0.2,node_color='red',node_radius=20)
    nx.draw_networkx_labels(G,pos,alpha=0.2,label_color='grey',font_size=12)
    plt.savefig("enron_graph.png")
    plt.show()
except Exception as e: # matplotlib not available
    print e
```

<h3>Defining functions to be used in features extraction (degree of centrality and number of connections in the node's graph</h3>

```python
bigG=nx.Graph() # create empty graph
for gr,vals in df.groupby(['fr','to']):
    bigG.add_edge(gr[0],gr[1],message=len(vals))
nx.write_gpickle(bigG,'/home/ipuser/bigG.pickle')
```

```python
%%writefile get_degree_centrality.py
def get_degree_centrality(input_data,field_name):
    import networkx as nx
    global bigG
    global deg_centr
    if bigG is None:
        bigG=nx.read_gpickle('/home/ipuser/bigG.pickle')
    if deg_centr is None:
        deg_centr = nx.degree_centrality(bigG)
    node = input_data[field_name].values[0]
    if node in deg_centr:
        return deg_centr[node]
    else:
        return 0.0

bigG = None
deg_centr = None
```

```python
from xpatterns.analytics.services.analytics_rest import AnalyticsRest
rest_call = AnalyticsRest()
rest_call.create_label_function('get_degree_centrality.py')
```

```python
%%writefile get_graph_connections.py
def get_graph_connections(input_data,field_name):
    import networkx as nx
    global bigG
    if bigG is None:
        bigG=nx.read_gpickle('/home/ipuser/bigG.pickle')
    node = input_data[field_name].values[0]
    if node in bigG.nodes():
        return len(nx.node_connected_component (bigG,node))
    else:
        return 0
```
