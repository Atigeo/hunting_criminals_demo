<h3>Input data (person's account transactions)</h3>

```python
from xpatterns.analytics.dal import DAL
d = DAL(use_shark=True)
trans = d.execute_query('SELECT date_time,user,amount FROM default.sim_trans_stream where big_trans>0 or is_in_odd_day>0 or is_at_odd_time>0;')
trans.head(10)
```

```python
def convert_to_float(x):
    result=0
    try:
        if '\n' in x:
            x= x.repalce('\n','')
        result=float(x)
    except:
        pass
    return result
def transform_date(date_string):
    if len(date_string)>8:
        date_string = date_string[:8]
    #print date_string
    result = datetime.datetime.strptime('19000101',"%Y%m%d")
    try:
        result = datetime.datetime.strptime(date_string,"%Y%m%d")
    except Exception as e:
        pass
    return result

cdf=trans.copy(deep=True)
cdf['amount']= np.array([convert_to_float(x) for x in trans['amount']])
cdf['datetime']= np.array([transform_date(x) for x in trans['date_time']])
cnts ={}
for user,ts in trans.groupby(['user']):
    cnts[user] = len(ts)
i =-1
for (v,k) in sorted([(v,k) for (k,v) in cnts.items()],reverse=True):
    i+=1
    if v>50 and i<10:
        print(k,v)
```

<h3>Detection of outlier transactions (OLS regression) </h3>

```python
from random import random
import statsmodels.graphics as smgraphics

import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.api import qqplot

for user,ts in cdf.groupby(['user']):
    cnts[user] = len(ts) 
    if len(ts)>60:
        ts.sort(columns=['date_time'],inplace=True)
        ts = ts.dropna()
        ts.index = pd.DatetimeIndex(ts.date_time)
        x= range(1,len(ts)+1)
        x= sm.add_constant(x)
        regression = sm.OLS(ts.amount.values,x).fit()
        # Find outliers #
        test = regression.outlier_test()
        
        #print test

        outliers = []
        for i in range(0,len(test)):
            if test[i][2]<0.5:
                outliers.append((i,test[i][2]))
        if len(outliers)>0:
            fig = plt.figure(figsize=(12,4))
            ts.amount.plot(title ='USER: '+user+' - '+str(len(ts))+' transactions')
            figure = smgraphics.regressionplots.plot_fit(regression,1)
            smgraphics.regressionplots.abline_plot(model_results = regression, ax=figure.axes[0])
            show()
            print 'OUTLIERS: ',outliers
```

<H3>Defining functions to be used in features extraction<h3>

```python
%%writefile count_outliers.py
def count_outliers(input_data,field_name,datetime_field):
    import statsmodels.api as sm
    import pandas as pd
    import numpy as np
    if len(input_data)<10:
        return 0
    cdf=input_data.copy(deep=True)
    cdf['amount']= np.array([convert_to_float(x) for x in input_data['amount']])
    cdf['datetime']= np.array([transform_date(x) for x in input_data['date_time']])
    cdf.sort(columns=[datetime_field],inplace=True)
    cdf.index = pd.DatetimeIndex(cdf[datetime_field])
    x= range(1,len(cdf)+1)
    x= sm.add_constant(x)
    regression = sm.OLS(cdf.amount.values,x).fit()
    test = regression.outlier_test()
    result = 0
    for i in range(0,len(test)):
        if test[i][2]<0.5:
            result += 1
    return result

def convert_to_float(x):
    result=0
    try:
        if '\n' in x:
            x= x.repalce('\n','')
        result=float(x)
    except:
        pass
    return result
def transform_date(date_string):
    import datetime
    if len(date_string)>8:
        date_string = date_string[:8]
    #print date_string
    result = datetime.datetime.strptime('19000101',"%Y%m%d")
    try:
        result = datetime.datetime.strptime(date_string,"%Y%m%d")
    except Exception as e:
        pass
    return result
```
