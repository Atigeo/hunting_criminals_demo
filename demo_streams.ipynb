```python
%matplotlib inline
import numpy as np
import datetime
from pylab import *
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
import IPython.html.widgets as widgets
from IPython.display import display, clear_output
import xpatterns.analytics.vizualization.viztools as viz
import time

pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
def convert_to_int(x):
    result=0
    try:
        result=int(x)
    except:
        pass
    return result

def convert_to_float(x):
    result=0
    try:
        result=float(x)
    except:
        pass
    return result
from xpatterns.analytics.dal import DAL
```

```python
!rm *.png
def draw_bar_chart(lst,file_name,ttl):
    import os
    fig = plt.figure(figsize=(6,4))
    plt.bar(range(1,len(lst)+1),lst,color='orange')
    title(ttl)
    try:
        os.remove(file_name)
    except Exception as e:
        pass
    xlabel('Iteration')
    ylabel('Count')
#     plt.show()
    fig.savefig(file_name)
    plt.close(fig)
draw_bar_chart([1230,1230,1230],'tst.png','Test')
```

```python
q = 'USE default; \n \
    drop table tmp_trans; \
    create table tmp_trans as \
    select user, sum(amount) as balance, std(amount) as std_amount,sum(is_in_odd_day) as odd_days, sum(is_at_odd_time) as odd_times from trans_stream \
    group by user order by user asc; \n \
    drop table tmp_email; \n \
    create table tmp_email as \
    select fr, sum(subject_sentiment) as subj_senti,sum(content_sentiment) as content_senti, \
    sum(subject_power) as subj_power,sum(content_power) as content_power, \
    sum(subject_topic) as subj_topic,sum(content_topic) as content_topic, \
    avg(fraud_score) as avg_f_s,std(fraud_score) as std_f_s \
    from email_stream \
    group by fr; \n\
    drop table tmp_feats; \n \
    create table tmp_feats as \
    select trans.*,email.* from tmp_trans trans \
    join tmp_email email on (trans.user = email.fr); \n \
    select * from tmp_feats \
    order by avg_f_s desc;'

emails_hist = []
transactions_hist = []
graph_edges_hist = []

step = 0

def iteration(a): 
    ! rm *.png
    lbl_transactions.value = ''
    lbl_emails.value = ''
    lbl_edges.value = ''
    global step
    global emails_hist
    global transactions_hist
    global graph_edges_hist
    
    step += 1
    form.description = 'Iteration '+str(step)+'- Counting records...'
    prg_wdg.value = 0
    prg_wdg.max = 4
    
    try:
        d = DAL(use_shark=True)
        transactions = int(d.execute_query('SELECT count(*) FROM default.trans_stream;'))
    except:
        transactions = 0

    transactions_hist.append(int(transactions))
    lbl_transactions.value = str(transactions)
    trans_img_f_n = 'trans'+str(step)+'.png'
    draw_bar_chart(transactions_hist,trans_img_f_n, 'PROCESSED TRANSACTIONS')

    try:
        d = DAL(use_shark=True)
        emails = int(d.execute_query('SELECT count(*) FROM default.email_stream;'))
    except:
        emails = 0
        
    emails_hist.append(int(emails))
    lbl_emails.value = str(emails)
    emails_img_f_n = 'emails'+str(step)+'.png'
    draw_bar_chart(emails_hist,emails_img_f_n,'PROCESSED EMAILS')
    
    try:
        d = DAL(use_shark=True)
        graph_edges = int(d.execute_query('SELECT count(*) FROM default.email_graph;'))
    except:
        graph_edges = 0

    graph_edges_hist.append(int(graph_edges))
    lbl_edges.value = str(graph_edges)
    edges_img_f_n = 'edges'+str(step)+'.png'
    draw_bar_chart(graph_edges_hist,edges_img_f_n,'PROCESSED GRAPH EDGES')
    
    prg_wdg.value = 1
    if emails<10 or transactions<10 or graph_edges<10:
        form.description = 'Iteration '+str(step)+' - Too few records...'
        prg_wdg.value = 4
        tbl_wdg.value ='<body><h3>Too few data points... </h3></body>'
    else:
        tbl_wdg.visible = False
        tbl_wdg.value = '<body><div><table><tr><td><img src="'+trans_img_f_n+'" width =400px height = 200px></img></td><td><img src="'+emails_img_f_n +'" width =400px height = 200px></img></td><td><img src="'+edges_img_f_n+'" width =400px height = 200px></img></td></tr></table></div>'
        tbl_wdg.visible = True
        
        form.description = 'Iteration '+str(step)+' - Generating features...'
        df = d.execute_query(q)
        prg_wdg.value = 2
        form.description = 'Iteration '+str(step)+' - Getting results...'
        df = d.execute_query('select * from tmp_feats order by avg_f_s desc;')
        if len(df.values)==0:
            return
        cdf=df.copy(deep=True)
        cdf['balance']= np.array([convert_to_float(x) for x in df['balance']])
        cdf['std_amount']= np.array([convert_to_float(x) for x in df['std_amount']])
        cdf['odd_days']= np.array([convert_to_int(x) for x in df['odd_days']])
        cdf['odd_times']= np.array([convert_to_int(x) for x in df['odd_times']])
        cdf['subj_senti']= np.array([convert_to_int(x) for x in df['subj_senti']])
        cdf['content_senti']= np.array([convert_to_int(x) for x in df['content_senti']])
        cdf['subj_power']= np.array([convert_to_int(x) for x in df['subj_power']])
        cdf['content_power']= np.array([convert_to_int(x) for x in df['content_power']])
        cdf['content_topic']= np.array([convert_to_int(x) for x in df['content_topic']])
        cdf['avg_f_s']= np.array([convert_to_float(x) for x in df['avg_f_s']])
        cdf['std_f_s']= np.array([convert_to_float(x) for x in df['std_f_s']])
        cdf.dropna()
#         clear_output()
        html = cdf[:5].to_html(justify='left')
        tbl_wdg.value += '<div>'+html+'</div>'
        
#         fig = plt.figure()
#         viz.plot_hist(cdf[['balance','odd_days','odd_times','subj_senti','content_senti','subj_power','content_power','content_topic','avg_f_s','std_f_s']],fig_cols=5)
#         fig.savefig('distr.png')
#         tbl_wdg.value +='<img src = "distr.png"></img><br>'
        fig =  plt.figure(figsize = (12,8))

        viz.plot_corr(cdf[['balance','odd_days','odd_times','subj_senti','content_senti','subj_power','content_power','content_topic','avg_f_s','std_f_s']], cmap='RdBu', mask_color='gray')
        matr_name = 'matr'+str(step)+'.png'
        try:
            os.remove(matr_name)
        except Exception as e:
            pass
        
        title('Iteration '+str(step) +' - Features correlation (Pearson)')
        fig.savefig(matr_name)
        plt.close(fig)
        tbl_wdg.visible = False
        tbl_wdg.value +='<div><img src = "'+matr_name+'" width=640px height =480px ></img></div></body>'
        tbl_wdg.visible = True
        form.description ='Iteration '+str(step)+ ' - rendered'
        prg_wdg.value = 4
#         clear_output()
        
#button = widgets.ButtonWidget(description = 'RUN ITERATION')
#button.on_click(iteration)

form = widgets.PopupWidget()
form.description = 'Features Extraction Iterations'

lbl_emails = widgets.TextWidget()
lbl_emails.value='0'
lbl_emails.description ='Emails processed'
lbl_emails.set_css({'width':'100px','text-align':'right'})
lbl_transactions = widgets.TextWidget()
lbl_transactions.value='0'
lbl_transactions.description =  'Transactions processed'
lbl_transactions.set_css({'width':'60px','text-align':'right'})
lbl_edges = widgets.TextWidget()
lbl_edges.value='0'
lbl_edges.description = 'Graph edges processed '
lbl_edges.set_css({'width':'60px','text-align':'right'})
labels_wdg = widgets.ContainerWidget()
labels_wdg.children = [lbl_transactions,lbl_emails,lbl_edges]

lbl_transactions.value = ''
lbl_emails.value = ''
lbl_edges.value = ''
charts_wdg = widgets.ContainerWidget()

tbl_wdg = widgets.HTMLWidget()
tbl_wdg.value = ''


charts_wdg.children = [tbl_wdg]

prg_wdg = widgets.IntProgressWidget()
prg_wdg.value = 0
prg_wdg.max = 4

form.children=[labels_wdg,prg_wdg,charts_wdg]
form.set_css({'width':'100%','height':'100%','bgcolor':'orange'})
display(form)
import time
for i in xrange(50):
    iteration(1)
    time.sleep(5)
```
